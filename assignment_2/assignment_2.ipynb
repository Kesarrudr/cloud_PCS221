```python
import random
import os

NUM_ROWS = 20000
NUM_COLS = 10000
file_path = '/content/drive/MyDrive/Cloud_Lab/large_matrix.csv'

print(f"Starting generation...")


directory = os.path.dirname(file_path)
if not os.path.exists(directory):
    os.makedirs(directory)
    print(f"Created directory: {directory}")


with open(file_path, 'w') as file:
    for row in range(NUM_ROWS):
        row_data = [];
        for col in range(NUM_COLS):
            if col < NUM_COLS / 3:
                row_data.append(str(random.randint(-50, 50)));
            elif col < 2 * NUM_COLS / 3:
                row_data.append(str(random.uniform(0.0, 5.0)))
            else:
                row_data.append(str(random.gauss(0, 1)))

        row_string = ','.join(row_data)

        file.write(row_string + "\n")

print("Data generation completed!")
```

```python
file_path = '/content/drive/MyDrive/Cloud_Lab/large_matrix.csv'
ROW_LIMIT = 100
matrix_dict = {}

with open(file_path, 'r') as file:
    first_row_line = file.readline().strip()
    headers = first_row_line.split(',')

    for col_num, val in enumerate(headers):
        unique_key = f"col_{col_num}_val_{val}"
        matrix_dict[unique_key] = []

    current_row = 1
    for line in file:
        if current_row >= ROW_LIMIT:
            break

        values = line.strip().split(',')

        for col_num, val in enumerate(values):
            unique_key = f"col_{col_num}_val_{headers[col_num]}"
            matrix_dict[unique_key].append(val)

        current_row += 1

first_key = list(matrix_dict.keys())[0]
print(f"Sample mapping for {first_key}: {matrix_dict[first_key][:5]}")
```

```python
file_path = '/content/drive/MyDrive/Cloud_Lab/large_matrix.csv'
ROW_LIMIT = 5
COL_LIMIT = 5
grouped_elements = {}

with open(file_path, 'r') as file:
    for row_idx, line in enumerate(file):
        if row_idx >= ROW_LIMIT:
            break

        values = line.strip().split(',')[:COL_LIMIT]

        for val in values:
            grouped_elements[val] = grouped_elements.get(val, 0) + 1

print("Grouped Elements (Element: Frequency):")
for element, count in grouped_elements.items():
    print(f"Element {element} appears {count} times")
```

```python
file_path = '/content/drive/MyDrive/Cloud_Lab/large_matrix.csv'
MATRIX_SIZE = 3
matrix_A = []
matrix_B = []

with open(file_path, 'r') as file:
    for row_idx, line in enumerate(file):
        if row_idx < MATRIX_SIZE:
            row_data = [int(float(x)) for x in line.strip().split(',')[:MATRIX_SIZE]]
            matrix_A.append(row_data)
        elif row_idx < MATRIX_SIZE * 2:
            row_data = [int(float(x)) for x in line.strip().split(',')[:MATRIX_SIZE]]
            matrix_B.append(row_data)
        else:
            break

matrix_addition = [
    [matrix_A[i][j] + matrix_B[i][j] for j in range(MATRIX_SIZE)]
    for i in range(MATRIX_SIZE)
]

matrix_subtraction = [
    [matrix_A[i][j] - matrix_B[i][j] for j in range(MATRIX_SIZE)]
    for i in range(MATRIX_SIZE)
]

print("Matrix A:", matrix_A)
print("Matrix B:", matrix_B)
print("Addition Result:", matrix_addition)
print("Subtraction Result:", matrix_subtraction)
```

```python
import random

N = 4
square_matrix = []

for i in range(N):
    row = []
    for j in range(N):
        even_number = random.randint(1, 50) * 2
        row.append(even_number)
    square_matrix.append(row)

for row in square_matrix:
    print(row)

sub_top_left = square_matrix[0][0]
sub_top_right = square_matrix[0][2]
sub_bottom_left = square_matrix[2][0]
sub_bottom_right = square_matrix[2][2]

sum_diag_1 = sub_top_left + sub_bottom_right
sum_diag_2 = sub_top_right + sub_bottom_left

print(f"\nTest Sub-matrix corners: TL={sub_top_left}, TR={sub_top_right}, BL={sub_bottom_left}, BR={sub_bottom_right}")
print(f"Diagonal 1 Sum: {sum_diag_1} (Even? {sum_diag_1 % 2 == 0})")
print(f"Diagonal 2 Sum: {sum_diag_2} (Even? {sum_diag_2 % 2 == 0})")
```

```python
file_path = '/content/drive/MyDrive/Cloud_Lab/large_matrix.csv'
ROW_LIMIT = 5
COL_LIMIT = 4

data_list = []
with open(file_path, 'r') as file:
    for row_idx, line in enumerate(file):
        if row_idx >= ROW_LIMIT:
            break

        row_tuple = tuple(int(float(x)) for x in line.strip().split(',')[:COL_LIMIT])
        data_list.append(row_tuple)

tuple_matrix = tuple(data_list)

row_sums = tuple(sum(row) for row in tuple_matrix)

print("Original Tuple Matrix:")
for row in tuple_matrix:
    print(row)

print("\nRow-wise Addition Results:")
for i, total in enumerate(row_sums):
    print(f"Sum of Row {i}: {total}")
```
